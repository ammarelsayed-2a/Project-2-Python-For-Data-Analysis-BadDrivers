# 🚗 Exploratory Data Analysis (EDA) — Bad Drivers Dataset
**Dataset:** `bad-drivers.csv` — FiveThirtyEight  
**Goal:** Explore fatal collision patterns across U.S. states and uncover key relationships between speeding, alcohol, insurance, and accident rates  
**Tools:** NumPy · Pandas · Matplotlib · Seaborn


## 1. Imports

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python
# Load the bad drivers dataset from CSV
data = pd.read_csv('bad-drivers.csv')
```

## 3. Rename Columns

```python
# The original column names are very long — renaming them for cleaner, more readable code
data.rename(columns={
    'Number of drivers involved in fatal collisions per billion miles'                        : 'fatal_collisions',
    'Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding'                   : 'pct_speeding',
    'Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired'           : 'pct_alcohol',
    'Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted'             : 'pct_not_distracted',
    'Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents': 'pct_no_prev_accidents',
    'Car Insurance Premiums ($)'                                                             : 'insurance_premium',
    'Losses incurred by insurance companies for collisions per insured driver ($)'           : 'insurance_losses'
}, inplace=True)

# Confirm new column names
print(data.columns.tolist())
```

## 7. Descriptive Statistics

```python
# Summary statistics for all numerical columns (mean, std, min, max, quartiles)
data.describe().T
```

## 8. Feature Engineering — Speed Level

```python
# Create a new categorical column 'speed_level' by binning pct_speeding into 3 groups
# This allows us to do categorical comparisons on a continuous variable
data['speed_level'] = pd.cut(
    data['pct_speeding'],
    bins=3,
    labels=['Low', 'Medium', 'High']
)

print(data['speed_level'].value_counts())
```

## 9. Univariate Analysis — Numerical Variables

```python
# Distribution of fatal collisions — are most states clustered around a specific rate?
sns.histplot(data['fatal_collisions'], bins=10, kde=True)
plt.title("Distribution of Fatal Collisions per Billion Miles")
plt.xlabel("Fatal Collisions")
plt.ylabel("Count")
plt.show()
```

```python
# Skewness tells us if the distribution leans left or right
print("Skewness:", data['fatal_collisions'].skew())
print("Mean:    ", data['fatal_collisions'].mean())
print("Median:  ", data['fatal_collisions'].median())
```

```python
# Boxplot to spot outliers in fatal collision rates
sns.boxplot(x=data['fatal_collisions'])
plt.title("Boxplot — Fatal Collisions per Billion Miles")
plt.show()
```

```python
# KDE plot of speeding percentage — smooth view of the distribution shape
sns.displot(data=data, x='pct_speeding', kind='kde')
plt.title("KDE — Speeding Percentage")
plt.show()
```

## 10. Univariate Analysis — Categorical Variables

```python
# How many states fall into each speed level category?
sns.countplot(x='speed_level', data=data, order=['Low', 'Medium', 'High'])
plt.title("Number of States per Speed Level")
plt.xlabel("Speed Level")
plt.ylabel("Count")
plt.show()
```

## 11. Bivariate Analysis — Numerical vs Numerical

```python
# Does higher speeding percentage lead to more fatal collisions?
sns.scatterplot(x='pct_speeding', y='fatal_collisions', data=data)
plt.title("Speeding % vs Fatal Collisions")
plt.show()
```

```python
# Add regression line to measure the strength of the relationship
sns.regplot(x='pct_speeding', y='fatal_collisions', data=data)
plt.title("Speeding % vs Fatal Collisions (Regression Line)")
plt.show()
```

## 12. Bivariate Analysis — Numerical vs Categorical

```python
# Do states with higher speeding levels also have more fatal collisions?
sns.boxplot(x='speed_level', y='fatal_collisions', data=data, order=['Low', 'Medium', 'High'])
plt.title("Fatal Collisions by Speed Level")
plt.show()
```

```python
# Average fatal collisions per speed level
sns.barplot(x='speed_level', y='fatal_collisions', data=data, order=['Low', 'Medium', 'High'])
plt.title("Average Fatal Collisions by Speed Level")
plt.show()
```

```python
# Violin plot combines distribution shape + summary stats in one view
sns.violinplot(x='speed_level', y='fatal_collisions', data=data, order=['Low', 'Medium', 'High'])
plt.title("Fatal Collisions Distribution by Speed Level (Violin)")
plt.show()
```

## 13. Multivariate Analysis

```python
# Each point = a U.S. state, colored by alcohol impairment level
# Explores speeding vs collisions while accounting for alcohol as a third dimension
sns.scatterplot(
    x='pct_speeding',
    y='fatal_collisions',
    hue='pct_alcohol',
    data=data
)
plt.title("Speeding vs Fatal Collisions (colored by Alcohol %)")
plt.show()
```

```python
# Same plot but colored by the engineered speed_level category
sns.scatterplot(
    x='pct_speeding',
    y='fatal_collisions',
    hue='speed_level',
    data=data
)
plt.title("Speeding vs Fatal Collisions (colored by Speed Level)")
plt.show()
```

```python
# KDE distribution of fatal collisions split by speed level
# Shows if the collision rate distribution differs across speed groups
sns.displot(
    data=data,
    x='fatal_collisions',
    hue='speed_level',
    kind='kde'
)
plt.title("Fatal Collisions KDE by Speed Level")
plt.show()
```

## 14. Correlation Analysis

```python
# Heatmap showing correlation between all numerical variables
# Values close to 1 or -1 = strong relationship; close to 0 = weak/no relationship
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

## 15. Pairplot — Key Variables Overview

```python
# Pairplot of the 3 most relevant variables — quick view of all pairwise relationships
sns.pairplot(data[['fatal_collisions', 'pct_speeding', 'pct_alcohol']])
plt.suptitle("Pairplot — Fatal Collisions, Speeding, Alcohol", y=1.02)
plt.show()
```

## 16. Key Insights

After analyzing the **Bad Drivers dataset** containing data from all 51 U.S. states:

1. **Speeding shows a weak positive correlation with fatal collisions**  
   Contrary to what one might expect, the correlation between speeding percentage and fatal collision rate is not very strong — suggesting other factors play a bigger role.

2. **Alcohol impairment is a notable factor**  
   States with higher alcohol-impaired driving percentages tend to cluster with higher fatal collision rates, making it one of the more impactful variables in the dataset.

3. **Fatal collision rates vary significantly by state**  
   The distribution is right-skewed — most states fall between 10–20 collisions per billion miles, but a few outliers (like North Dakota and South Carolina) exceed 23.

4. **Insurance premiums don't strongly predict collision rates**  
   Higher insurance premiums do not always correspond to higher accident rates — premium pricing is likely influenced by factors beyond just collision frequency.

5. **States with medium speeding levels show the widest spread in fatal collisions**  
   The violin plot reveals that medium-speed states have the most variance, while high-speed states are more tightly clustered — suggesting speeding alone isn't the key driver.

6. **Insurance losses and premiums are moderately correlated**  
   States where insurance companies incur higher losses also tend to charge higher premiums — an expected but confirmed relationship in the data.
